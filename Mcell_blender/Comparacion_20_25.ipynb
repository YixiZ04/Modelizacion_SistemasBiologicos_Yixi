```python
# To create a list with path to data directories
data_dirs_paths_20 = []
seed_numbers_20 = []
for num in range (1,11,1):
    seed_num = str(num)
    while len (seed_num) < 5:
        seed_num = "0" + seed_num
    path_2_dir = input_path_20 + seed_num + "/"
    data_dirs_paths_20.append(path_2_dir)
    seed_numbers_20.append(seed_num)
data_dirs_paths_20
```

```python
# To create a list with path to data directories
data_dirs_paths_25 = []
seed_numbers_25 = []
for num in range (1,11,1):
    seed_num = str(num)
    while len (seed_num) < 5:
        seed_num = "0" + seed_num
    path_2_dir = (input_path_25+ seed_num + "/")
    data_dirs_paths_25.append(path_2_dir)
    seed_numbers_25.append(seed_num)
data_dirs_paths_25
```

```python
# The files that we have in every directory
files = ["glu.Cube.dat", "glu.World.dat", "O.PSD.dat"]
files
```

```python
# Fetch data from every directory. Here every data is a list in a global list
cube_data_20 = []
world_data_20 = []
channel_data_20 = []
time = []
for dir_path in data_dirs_paths_20:
    index = 0
    for file in files:
        index += 1
        filename = dir_path + file
        f = [ line.rstrip ('\n') for line in open (filename).readlines() ]
        time = [ float (line.split (' ') [0]) for line in f]
        data = [ int (line.split (' ') [1]) for line in f]
        if index == 1:
            cube_data_20.append (data)
        elif index == 2:
            world_data_20.append (data)
        else:
            channel_data_20.append (data)
len (time), len (cube_data_20), len(world_data_20), len(channel_data_20)
# Now seed_0 data is the 0th item in each list, and seed_1 is the 1st item...
```

```python
# Fetch data from every directory. Here every data is a list in a global list
cube_data_25 = []
world_data_25 = []
channel_data_25 = []
time = []
for dir_path in data_dirs_paths_25:
    index = 0
    for file in files:
        index += 1
        filename = dir_path + file
        f = [ line.rstrip ('\n') for line in open (filename).readlines() ]
        time = [ float (line.split (' ') [0]) for line in f]
        data = [ int (line.split (' ') [1]) for line in f]
        if index == 1:
            cube_data_25.append (data)
        elif index == 2:
            world_data_25.append (data)
        else:
            channel_data_25.append (data)
len (time), len (cube_data_25), len(world_data_25), len(channel_data_25)
# Now seed_0 data is the 0th item in each list, and seed_1 is the 1st item...
```

```python
# Calculate media from each data file

def calculate_mean (list_of_data):
    means = []
    num_replicates = len(list_of_data)
    num_data = len (list_of_data [0])
    for i in range (num_data):
        mean = 0
        for j in range (num_replicates):
            mean += (list_of_data[j][i])
        mean  = mean / num_replicates
        means.append(mean)
    return means
```

```python
## Opened channel meanc
channel_mean_20 = calculate_mean(channel_data_20)
plt.plot (time, channel_mean_20, label = "channel_mean")
plt.xlim ([0, 0.003])
plt.grid (True)
plt.legend ()
plt.show ()
```

```python
# Get max number of channel opened
def get_curve_peak (x_ax, y_ax):
    y_peak = max (y_ax)
    max_index = y_ax.index (y_peak)
    x_peak = x_ax [max_index]
    return x_peak, y_peak
```

```python
time_peak_20, channels_peak_20 = get_curve_peak(time, channel_mean_20)
time_peak_25, channels_peak_25 = get_curve_peak(time, channel_mean_25)
point_20 = [time_peak_20, channels_peak_20]
point_25 = [time_peak_25, channels_peak_25]
```

```python
channel_per_data_mean_20 = calculate_mean(channel_per_data_20)
channel_per_data_mean_25 = calculate_mean(channel_per_data_25)
plt.plot (time, channel_per_data_mean_20, label = "channel_per_data_mean_20")
plt.plot (time, channel_per_data_mean_25, 'r', label = "channel_per_data_mean_25")
plt.legend ()
plt.show ()
```

```python
NA = 6.022e23
vol_cube = 4.029e-3  #mum^3
vol_world = 6  # mum^3
vol_cube_L = vol_cube / 1e15
vol_world_L = vol_world / 1e15


def convert_molar_concentration(num_molecules, volume, avogadro=NA):
    return (num_molecules / avogadro) / volume
```

```python
cube_mean_20 = calculate_mean(cube_data_20)
cube_mean_25 = calculate_mean(cube_data_25)
mean_molar_20 = []
mean_molar_25 = []
for data in cube_data_20:
    temp_list = []
    for mol_number in data:
        molar_concentration = convert_molar_concentration(mol_number, vol_cube_L)
        temp_list.append(molar_concentration)
    mean_molar_20.append(temp_list)
molar_cube_25 = []
for data in cube_data_25:
    temp_list = []
    for mol_number in data:
        molar_concentration = convert_molar_concentration(mol_number, vol_cube_L)
        temp_list.append(molar_concentration)
    mean_molar_25.append(temp_list)

mean_molar_20 = calculate_mean(mean_molar_20)
mean_molar_25 = calculate_mean(mean_molar_25)
```

```python
## Important plot

fig, ax1 = plt.subplots (figsize = (8,7))
ax1.plot (time, cube_mean_20, 'b--',label = "HS = 20 nm")
ax1.plot (time, cube_mean_25, 'r', label = "HS = 25 nm")
ax1.grid (True)
ax1.set_xlabel('Tiempo (s)')
ax1.set_ylabel ('Nmero de molculas Glu')
ax1.legend ()
ax2 = ax1.twinx ()
ax2.plot (time, mean_molar_20, 'b--')
ax2.plot (time, mean_molar_25, 'r')
ax2.set_ylabel('Concentracin molar (mol/L)')

plt.xlim ([0, 0.003])
plt.title ("Comparacin Glu HS 20 vs 25 nm")

saving_figures ("Comparacin Glu 20_25 HS")
plt.show ()
```
